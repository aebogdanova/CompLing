С помощью регулярных выражение очистите текст от тэгов (мета информацию внутри тэгов тоже удалите). В получившемся тексте точки не должны слипаться с началом следующего предложений, также не должно быть лишних пробелов.

```python
import re
```

```python
def remove_tags(text):

    # find substrings with tags in angle brackets (using non-greedy qualifier *?) and replace them with space character
    text = re.sub('<.*?>', ' ', text)
    # find substrings of more than one space character and replace them with one
    text = re.sub(' +', ' ', text)
    # delete space characters followed by punctuation characters (.,!?\n) (using positive lookahead assertion ?=)
    text = re.sub(' +(?=\.|,|!|\?|\n)', '', text)

    return text
```

## 4. Pymorphy vs Mystem (2 балла)

Лемматизируйте очищенный текст с помощью mystem и pymorphy. Сравните результаты: есть ли разница, что лучше. 

Для пайморфи используйте токенизацию из mystem, чтобы исключить влияние токенизации на результат.

```python
import pandas as pd
from pymystem3 import Mystem
from pymorphy2 import MorphAnalyzer
mystem = Mystem()
pymorphy = MorphAnalyzer()
```

```python
# токенизируем текст на предложения простым сплитом (понадобится для дальнейшего сравнения работы морфоанализаторов)
text_tokenized = re.split('(?<=[\.!?])\s(?=[А-Я0-9])', text)
```

```python
# сохраняем в словарь слова (с грамматической характеристикой и предложением, в котором они встретились), 
# которые по-разному лемматизировались морфоанализаторами
analysis_dict = {}
analysis_count = 0
for sent in text_tokenized:
    sent_lemmatized_mystem = mystem.lemmatize(sent)
    sent_analyzed_mystem = mystem.analyze(sent)
    sent_analyzed_pymorphy = [pymorphy.parse(word['text'])[0] for word in sent_analyzed_mystem]
    for x, y, z in zip(sent_lemmatized_mystem, sent_analyzed_mystem, sent_analyzed_pymorphy):
        if x != z.normal_form:
            try:
                analysis_dict[analysis_count] = {
                    'token': y['text'],
                    'context': sent,
                    'lemma_mystem': x,
                    'gr_mystem': y['analysis'][0]['gr'],
                    'lemma_pymorphy': z.normal_form,
                    'gr_pymorphy': z.tag.cyr_repr
                }
                analysis_count += 1
            except:
                analysis_dict[analysis_count] = {
                    'token': y['text'],
                    'context': sent,
                    'lemma_mystem': x,
                    'gr_mystem': 'None', # для тех лемм, у которых POS в mystem не определился
                    'lemma_pymorphy': z.normal_form,
                    'gr_pymorphy': z.tag.cyr_repr
                }
                analysis_count += 1
print('Number of mismatched lemmas:', len(analysis_dict))
            
```

```python
# выводим несовпавшие леммы таблицей
analysis = pd.DataFrame.from_dict(analysis_dict, orient = 'index')
pd.options.display.max_rows = None
pd.set_option('display.max_colwidth', None)
display(analysis)
```

Одна из фичей майстема - контекстная дизамбигуация (т.е. определение правильного разбора в зависимости от контекста).

Запустите mystem на очищенном тексте с дизамбигуацией и без и проверьте есть ли какие-то различия.

```python
# проводим морфоанализ с дизамбигуацией и без (для снятия омонимии режим вывода -d)
# /можно было бы проводить анализ через pymystem с параметром disambiguation=False, но нам понадобится разделение на предложения,
# чтобы сравнивать работу mystem со снятием омонимии и без, поэтому будем анализировать напрямую через mystem из командной строки/

!mystem -isc --format json text.txt text_ambiguated.txt
text_ambiguated = [json.loads(line) for line in open('text_ambiguated.txt', encoding='utf-8')]

!mystem -iscd --format json text.txt text_disambiguated.txt
text_disambiguated  = [json.loads(line) for line in open('text_disambiguated.txt', encoding='utf-8')]
```

```python
# для дальнейшего сравнения анализов с дизамбигуацией и без понадобится разбиение на предложения
def sent_tokenize(text):
    all_sents = []
    for paragraph in text:
        sep = [token['text'] for token in paragraph if token['text'] == '\\s']
        sents = [[] for _ in range(len(sep)+1)]
        count = 0
        for token in paragraph:
            while token['text'] != '\\s':
                sents[count].append(token)
                break
            else:
                count += 1
                continue
        for sent in sents:
            all_sents.append(sent)
    return all_sents
```

```python
# разбиваем проанализированные тексты на предложения
text_amb_tokenized = sent_tokenize(text_ambiguated)
text_disamb_tokenized = sent_tokenize(text_disambiguated)
```

```python
# сохраняем в словарь те слова, которые проанализировались по-разному (вместе с грамматической разметкой и предложением)
disambiguation_dict = {}
disambiguation_count = 0
for sent_amb, sent_disamb in zip(text_amb_tokenized, text_disamb_tokenized):
    for token_amb, token_disamb in zip(sent_amb, sent_disamb):
        try:
            if token_amb['analysis'][0] != token_disamb['analysis'][0]:
                disambiguation_dict[disambiguation_count] = {
                    'token': token_amb['text'],
                    'context': ''.join([token['text'] for token in sent_amb]),
                    'lex_amb': token_amb['analysis'][0]['lex'], 
                    'gr_amb': token_amb['analysis'][0]['gr'], 
                    'lex_disamb': token_disamb['analysis'][0]['lex'], 
                    'gr_disamb': token_disamb['analysis'][0]['gr']
                }
                disambiguation_count += 1
        except:
            pass
print('Number of differences:', len(disambiguation_dict))    
```

Mystem со снятием омонимии работает несколько лучше: <br>
1. Улавливаются адъективированные причастия: в контексте "*таким говорящим именем*" mystem с дизамбигуацией *говорящим* определяет как прилагательное (*говорящий_A*), а без нее - как причастие с леммой *говорить*.<br><br>
2. Лучше распознаются различия между местоимениями/наречиями и союзами: например, в контексте "*Что это такое?*" анализатор со снятием омонимии верно определяет *что* как местоимение-существительное (*что_SPRO*), тогда как без снятия омонимии оно определялось как союз; в контексте "*чем активнее область, тем она «теплее»*" токен *тем* верно определяется как союз (*тем_CONJ*), а без снятия омонимии - как местоимение (*то_SPRO*). Однако ошибки даже при снятой омонимии все равно встречаются: например, *как* в контексте "*Как я уже упоминал...*" определен как местоименное наречие, а не как союз. Также ошибка встретилась, например, в контексте "*Сколько это стоит?*": токену *сколько* mystem со снятием омонимии присвоил часть речи союз (*сколько_CONJ*), при этом без снятой омонимии часть речи определяется верно (*сколько_ADV*).<br><br>
3. Лучше находятся числительные. Например, без дизамбигуации для токена *одну* в контексте *4 платных линии и одну бесплатную* mystem присваивает часть речи местоимение-прилагательное (*один_APRO*), а с диазмбигуацией - уже числительное-прилагательное (*один_ANUM*). Здесь также стоит отметить, что даже при ошибочном написании *пол года* в контексте "*популяризировался пол года-год назад*" mystem со снятием омонимии *пол* относит к числительным (*пол_NUM*), тогда как без снятия омонимии *пол* анализируется как существительное мужского рода (*пол_S*).<br><br>
4. Еще несколько ярких примеров, когда обращение к соседним словам улучшает морфологический анализ: 1) в контексте "*для тестирования нового стартапа*" дизамбигуация позволяет верно определить неодушевленность для токена *стартапа* (*стартап_S,муж,неод*, без дизамбигуации - *стартапа_S,имя,од*); 2) в контексте "*Стоит она 19 долларов...*" для токена *стоит* дизамбигуация верно выявляет лемму *стоить_V* (без дизамбигуации - *стоять_V*) - здесь отдельно отметим, что для других контекстов этот же глагол лемматизировался неправильно (как описано в задании 4); 3) в контексте "*Да-да, сервис не бесплатен*" снятие омонимии помогает отнести *да* к частицам (*да_PART*); 4) в контексте "*Всех с прошедшим Новым годом!*" *прошедшее* верно определяется как прилагательное (*прошедший_A*, без дизамбигуации - как существительное *прошедшее_S*).

Несмотря на указанные выше примеры, демонстрирующие эффективность снятия омонимии для mystem, нельзя однозначно сказать, что во всех случаях дизамбигуация приводит к улучшению морфоанализа. В некоторых контекстах ошибки все равно встречаются, в том числе тогда, когда без снятия омонимии морфоанализатор определил часть речи и лемму верно.
